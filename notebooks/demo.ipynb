# RAG Sufficient Context — Interactive Demo

Loads a precomputed experiment run and lets you slide a score threshold to see live coverage/risk/selective accuracy.

**Prerequisites:** run `python -m run.pipeline.run_experiment configs/experiments/qwen7b_hybrid.yaml` (or any other config) at least once so that `selective_curves.json` exists.

```python
import json
from pathlib import Path

experiment = 'qwen7b_hybrid'
curves_path = Path('..') / 'results' / experiment / 'selective_curves.json'
with open(curves_path) as f:
    curves = json.load(f)

scores = curves['gate_scores']
labels = curves['labels']
total = len(labels) + curves.get('num_abstained', 0) or len(labels)
print(f'loaded {len(scores)} answered examples')
```

```python
import sys
sys.path.insert(0, '..')
from src.demo import build_threshold_widget

build_threshold_widget(scores=scores, labels=labels, total=total)
```
